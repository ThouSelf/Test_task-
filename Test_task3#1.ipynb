```python
import pandas as pd
import numpy as np

from sklearn.metrics import mean_squared_error
```

```python
df_train = pd.read_csv('internship_train.csv')

df_test = pd.read_csv('internship_hidden_test.csv')
```

```python
df_train
```

```python
X_train = df_train.iloc[:, :53]
y_train = df_train.iloc[:, 53:]
y_train
X_train
```

```python
X_test = df_test
```

```python
mean = X_train.mean(axis = 0)
std = X_train.mean(axis = 0)
X_train -= mean
X_train /= std
X_test -= mean
X_test /= std
X_train
```

```python

from keras.models import Sequential
from keras.layers import Dense
from keras.callbacks import ModelCheckpoint 

model = Sequential()
model.add(Dense(128, activation = 'relu', input_shape=(X_train.shape[1],)))
model.add(Dense(1))
model.compile(optimizer = 'adam', loss = 'mse', metrics = ['mse'])
```

```python
filename = 'Weights_Test_Task3.hdf5'
checkpointer = ModelCheckpoint(filepath = filename, verbose = 1, save_weights_only = False)
model.fit(X_train, y_train, epochs = 30, callbacks = [checkpointer], batch_size = 128, verbose = 2)
```

```python
predicted = model.predict(X_test)
predicted
```
